```python
df["Ram"]=df["Ram"].astype(int)
```

```python
df["Weight"]=df["Weight"].str.replace("kg","")
```

```python
df['Weight'] = df['Weight'].str.replace("?", "")
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
```

```python
df["Weight"]=df["Weight"].astype(float)
```

```python
df['Weight'].fillna(df["Weight"].mean(), inplace=True)
```

```python
X=df.drop(columns=['Price'])
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
X.isnull().sum()
```

```python
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(handle_unknown="ignore",sparse=False,drop='first'),[0,1,3,4,6,7,8])
],remainder='passthrough')
step2=LinearRegression()
pipe=Pipeline([
    ('step1',step1),
    ('step2',step2)
])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(handle_unknown="ignore",sparse=False,drop='first'),[0,1,3,4,6,7,8])
],remainder='passthrough')
step2=XGBRegressor(n_estimators=45,max_depth=5,learning_rate=0.5)
pipe=Pipeline([
    ('step1',step1),
    ('step2',step2)
])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
def predict_price(Company,TypeName,Inches,ScreenResolution,Cpu,Ram,Memory,Gpu,OpSys,Weight):
  result=round(pipe.predict([[Company,TypeName,Inches,ScreenResolution,Cpu,Ram,Memory,Gpu,OpSys,Weight]])[0],2)
  print(f"The estimated Laptop price is {result} Rupees")
```

```python
predict_price("HP","Notebook",15.6,"Full HD 1920x1080","Intel Core i7 7600U 2.8GHz	",16,"256GB SSD","Nvidia GeForce 930MX","Windows 10",1.93)
```
